```python
import pandas as pd
import numpy as np
from lazypredict.Supervised import LazyRegressor
```

```python
df = pd.read_csv("TCS with all features.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
df.head()
```

```python
x = df.loc[:, df.columns != "Next day close price"]
y = df.loc[:, df.columns == "Next day close price"]
```

```python
x.shape, y.shape
```

```python
split = int(0.8 * len(x))
```

```python
x_train, x_test, y_train, y_test = np.array(x[:split]),np.array(x[split:]),np.array(y[:split]),np.array(y[split:])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(x_train)

# Apply transform to both the training set and the test set.
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(.95)

pca.fit(x_train)

x_train = pca.transform(x_train)
x_test = pca.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
clf = LazyRegressor(ignore_warnings = True)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
```
